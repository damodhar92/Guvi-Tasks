**Importing the libraries**

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split      
from sklearn.tree import DecisionTreeClassifier
```

**Importing the dataset**

```python
df=pd.read_csv('/content/Social_Network_Ads.csv')
```

**Splitting the dataset into the Training set and Test set**

```python
x=df[["Age", "EstimatedSalary"]]
y=df["Purchased"]
x_train,  x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=5)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

**Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
func = StandardScaler()
func.fit(x_train)
x_train_std = func.transform(x_train)
x_test_std = func.transform(x_test)
```

**Fitting Decision Tree Classification to the Training set**

```python
model = DecisionTreeClassifier()
model.fit(x_train_std,y_train)
model.score(x_test_std,y_test)
     
```

**Predicting the Test set results**

```python
y_pred = model.predict(x_test_std)
```

**Making the Confusion Matrix**

```python
from sklearn.metrics import confusion_matrix
cmatrix = confusion_matrix(y_test, y_pred) 
   
sns.heatmap(cmatrix, annot=True)  
```

**Visualising the Training set results**

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

xs = x_train_std[:,0]
ys = x_train_std[:,1]
zs = y_train
ax.scatter(xs, ys, zs, s=50,color=['blue' if p==1 else 'green' for p in model.predict(x_train_std)], alpha=0.6, edgecolors='w')

ax.set_xlabel('Age')
ax.set_ylabel('EstimatedSalary')
ax.set_zlabel('Purchased')
```

**Visualising the Test set results**

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

xs = x_test_std[:,0]
ys = x_test_std[:,1]
zs = y_test
ax.scatter(xs, ys, zs, s=50,color=['blue' if p==1 else 'green' for p in y_pred], alpha=0.6, edgecolors='w')

ax.set_xlabel('Age')
ax.set_ylabel('EstimatedSalary')
ax.set_zlabel('Purchased')
```
